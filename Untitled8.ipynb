# Input Sentence

     |
Encoder LSTM
     |
Final Hidden State + Cell State
     |
Context Vector
     |
Decoder LSTM
     |
Generate Output Word-by-Word

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras. layers import Input, LSTM, Dense,Embedding
from tensorflow.keras.preprocessing. text import Tokenizer
from tensorflow.keras.preprocessing. sequence import pad_sequences
```

Prepare Data

```python
data = [
    ("i am learning", "je suis en train d apprendre"),
    ("he is running", "il court"),
    ("she is happy", "elle est heureuse"),
    ("i am happy", "je suis heureux")
]
input_texts=[x[0] for x in data]
# Add spaces around <start> and <end> to ensure they are tokenized as separate words
target_texts=["<start> " + x[1] + " <end>" for x in data]
print(input_texts)
print(target_texts)
```

# Tokenization

```python
# Encoder tokenizer
input_tokenizer = Tokenizer()
input_tokenizer.fit_on_texts(input_texts)
input_sequences = input_tokenizer.texts_to_sequences(input_texts)

# Decoder tokenizer
target_tokenizer = Tokenizer(filters='')
target_tokenizer.fit_on_texts(target_texts)
target_sequences = target_tokenizer.texts_to_sequences(target_texts)
```

# Padding

```python
max_input_len = max(len(seq) for seq in input_sequences)
max_target_len = max(len(seq) for seq in target_sequences)

encoder_input_data = pad_sequences(input_sequences, maxlen=max_input_len, padding='post')
decoder_input_data = pad_sequences(target_sequences, maxlen=max_target_len, padding='post')
```

# Create Decoder Target(Shifted)

decoder input=...
decoder output=....

```python
decoder_target_data = np.zeros_like(decoder_input_data)

for i in range(len(data)):
    decoder_target_data[i, :-1] = decoder_input_data[i, 1:]
```

# Build Model
Parameters

```python
vocab_size_input = len(input_tokenizer.word_index) + 1
vocab_size_target = len(target_tokenizer.word_index) + 1
embedding_dim = 50
latent_dim = 100
```

# Encoder

```python
encoder_inputs=Input(shape=(None,))
enc_emb=Embedding(vocab_size_input,embedding_dim)(encoder_inputs)

encoder_lstm=LSTM(latent_dim,return_state=True)
_, state_h, state_c = encoder_lstm(enc_emb)
#Encoder understands meaning and stores it in state_h, start_c, These togeather from Context vector
encoder_states = [state_h, state_c]
print(encoder_states)
```

# Decoder

```python
decoder_inputs = Input(shape=(None,))
dec_emb = Embedding(vocab_size_target, embedding_dim)(decoder_inputs)

decoder_lstm = LSTM(latent_dim, return_sequences=True, return_state=True)
decoder_outputs, _, _ = decoder_lstm(dec_emb, initial_state=encoder_states)

decoder_dense = Dense(vocab_size_target, activation='softmax')
decoder_outputs = decoder_dense(decoder_outputs)
```

# Compile Model

```python
model = Model([encoder_inputs, decoder_inputs], decoder_outputs)
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy')
```

Train Model

```python
model.fit(
    [encoder_input_data, decoder_input_data],
    decoder_target_data,
    batch_size=2,
    epochs=200
)
```

# Inference Model

Training model # Prediction model

We build separate inference models

# Encoder Interface

```python
encoder_model=Model(encoder_inputs,encoder_states)
```

Decoder Interface

```python
decoder_state_input_h = Input(shape=(latent_dim,))
decoder_state_input_c = Input(shape=(latent_dim,))
decoder_states_inputs = [decoder_state_input_h, decoder_state_input_c]

dec_emb2 = Embedding(vocab_size_target, embedding_dim)(decoder_inputs)

decoder_outputs2, state_h2, state_c2 = decoder_lstm(
    dec_emb2, initial_state=decoder_states_inputs
)

decoder_states2 = [state_h2, state_c2]
decoder_outputs2 = decoder_dense(decoder_outputs2)

decoder_model = Model(
    [decoder_inputs] + decoder_states_inputs,
    [decoder_outputs2] + decoder_states2
)
```

```python
reverse_target_index = {i: word for word, i in target_tokenizer.word_index.items()}

def decode_sequence(input_seq):

    states_value = encoder_model.predict(input_seq)

    # <start> and <end> are now expected to be properly tokenized due to spaces in target_texts
    start_token = target_tokenizer.word_index['<start>']
    end_token = target_tokenizer.word_index['<end>']

    target_seq = np.array([[start_token]])

    stop_condition = False
    decoded_sentence = ""

    while not stop_condition:

        output_tokens, h, c = decoder_model.predict(
            [target_seq] + states_value
        )

        sampled_token_index = np.argmax(output_tokens[0, -1, :])

        sampled_word = reverse_target_index.get(
            sampled_token_index, ''
        )

        if (
            sampled_token_index == end_token or
            len(decoded_sentence.split()) > max_target_len
        ):
            stop_condition = True
        else:
            decoded_sentence += " " + sampled_word

        target_seq = np.array([[sampled_token_index]])

        states_value = [h, c]

    return decoded_sentence
```

# Task2:Text Preprocessing

```python
#Lowercase Conversion
df['text'] = df['text'].str.lower()
df['headlines'] = df['headlines'].str.lower()
```

```python
#punctuation removal
import re

def clean_text(text):
    text = re.sub(r'[^a-zA-Z\s]', '', str(text))
    return text

df['text'] = df['text'].apply(clean_text)
df['headlines'] = df['headlines'].apply(clean_text)
```

```python
#addding start and end tokens
df['headlines'] = df['headlines'].apply(
    lambda x: '<start> ' + x + ' <end>'
)
```

# Tokenization

```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

# Encoder Tokenization

```python
encoder_tokenizer = Tokenizer()

encoder_tokenizer.fit_on_texts(df['text'])

encoder_sequences = encoder_tokenizer.texts_to_sequences(df['text'])
```

# Decoder Tokenizer

```python
decoder_tokenizer = Tokenizer(filters='')

decoder_tokenizer.fit_on_texts(df['headlines'])

decoder_sequences = decoder_tokenizer.texts_to_sequences(df['headlines'])
```

```python
#vocabulary sizes
encoder_vocab_size = len(encoder_tokenizer.word_index) + 1

decoder_vocab_size = len(decoder_tokenizer.word_index) + 1

print("Encoder Vocab Size:", encoder_vocab_size)
print("Decoder Vocab Size:", decoder_vocab_size)
```

# Padding

```python
max_text_len = max(len(seq) for seq in encoder_sequences)

max_summary_len = max(len(seq) for seq in decoder_sequences)

print(max_text_len)
print(max_summary_len)
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

encoder_input_data = pad_sequences(
    encoder_sequences,
    maxlen=max_text_len,
    padding='post'
)

decoder_input_data = pad_sequences(
    decoder_sequences,
    maxlen=max_summary_len,
    padding='post'
)
```

Create Decoder Target data

```python
import numpy as np

decoder_target_data = np.zeros_like(decoder_input_data)

for i in range(len(df)):
    decoder_target_data[i, :-1] = decoder_input_data[i, 1:]
```

```python
#final shapes
print("Encoder Input Shape:", encoder_input_data.shape)
print("Decoder Input Shape:", decoder_input_data.shape)
print("Decoder Target Shape:", decoder_target_data.shape)
```

# Task3:Build Encoder

```python
from tensorflow.keras.layers import Input, Embedding, LSTM
from tensorflow.keras.models import Model

latent_dim = 256

# Encoder Input
encoder_inputs = Input(shape=(max_text_len,))

# Embedding Layer
encoder_embedding = Embedding(
    input_dim=encoder_vocab_size,
    output_dim=128,
    mask_zero=True
)(encoder_inputs)

# Encoder LSTM
encoder_lstm = LSTM(
    latent_dim,
    return_state=True
)

encoder_outputs, state_h, state_c = encoder_lstm(encoder_embedding)

# Context Vector
encoder_states = [state_h, state_c]
```

# Build Decoder

```python
from tensorflow.keras.layers import Dense

# Decoder Input
decoder_inputs = Input(shape=(max_summary_len,))

# Decoder Embedding
decoder_embedding = Embedding(
    input_dim=decoder_vocab_size,
    output_dim=128,
    mask_zero=True
)(decoder_inputs)

# Decoder LSTM
decoder_lstm = LSTM(
    latent_dim,
    return_sequences=True,
    return_state=True
)

decoder_outputs, _, _ = decoder_lstm(
    decoder_embedding,
    initial_state=encoder_states
)

# Dense Layer
decoder_dense = Dense(
    decoder_vocab_size,
    activation='softmax'
)

decoder_outputs = decoder_dense(decoder_outputs)
```

```python
from tensorflow.keras.models import Model

model = Model(
    [encoder_inputs, decoder_inputs],
    decoder_outputs
)

model.summary()
```

# Task 5: Train Seq2Seq Model

```python
from tensorflow.keras.utils import to_categorical

decoder_target_data = to_categorical(
    decoder_target_data,
    num_classes=decoder_vocab_size
)
```

# Compile model

```python
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

# Train model

```python
history = model.fit(
    [encoder_input_data, decoder_input_data],
    decoder_target_data,
    batch_size=64,
    epochs=10,
    validation_split=0.2
)
```

# Task6:Generate summary

```python
#build inferecnce encoder
encoder_model = Model(
    encoder_inputs,
    encoder_states
)

#build inference decoder
from tensorflow.keras.layers import Input

decoder_state_input_h = Input(shape=(latent_dim,))
decoder_state_input_c = Input(shape=(latent_dim,))

decoder_states_inputs = [
    decoder_state_input_h,
    decoder_state_input_c
]

decoder_emb2 = decoder_embedding

decoder_outputs2, state_h2, state_c2 = decoder_lstm(
    decoder_emb2,
    initial_state=decoder_states_inputs
)

decoder_states2 = [state_h2, state_c2]

decoder_outputs2 = decoder_dense(decoder_outputs2)

decoder_model = Model(
    [decoder_inputs] + decoder_states_inputs,
    [decoder_outputs2] + decoder_states2
)

#reverse ken dictionary
reverse_target_word_index = {
    i: word
    for word, i in decoder_tokenizer.word_index.items()
}

target_word_index = decoder_tokenizer.word_index

#summary generation function
import numpy as np

def decode_sequence(input_seq):

    states_value = encoder_model.predict(input_seq, verbose=0)

    target_seq = np.zeros((1,1))

    target_seq[0,0] = target_word_index['start']

    stop_condition = False
    decoded_sentence = ""

    while not stop_condition:

        output_tokens, h, c = decoder_model.predict(
            [target_seq] + states_value,
            verbose=0
        )

        sampled_token_index = np.argmax(output_tokens[0,-1,:])

        sampled_word = reverse_target_word_index.get(
            sampled_token_index,
            ''
        )

        if sampled_word != 'end':
            decoded_sentence += sampled_word + ' '

        if (sampled_word == 'end' or
            len(decoded_sentence.split()) > max_summary_len):

            stop_condition = True

        target_seq = np.zeros((1,1))
        target_seq[0,0] = sampled_token_index

        states_value = [h,c]

    return decoded_sentence
```
